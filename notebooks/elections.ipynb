### <font color=steelblue> 1.3 - Funções

#### <font color=slategray> 1.3.1 - plot_gasto_vs_votos()

**O que faz:** scatter de gasto vs votos por situação eleitoral com 3 grupos de pontos (Eleito, Não Eleito, Suplente), 4 linhas verticais de referência (mediana e média de gasto por grupo) e escala logarítmica nos dois eixos.

**Por que existe:**
- logica complexa (3 grupos, multiplas linhas de referancia) que tornaria cada celula de uso ilegível
- chamada 4 vezes com a mesma estrutura: Brasil, RJ, GO e AM
- o parâmetro `estados` permite filtrar sem duplicar código

```python
def plot_gasto_vs_votos(df, estados=None):
    """
    Args:
        df      : DataFrame — eleicoes_enriquecido ou qualquer subconjunto
        estados : list[str] ou None — ex: ['SP'], ['SP', 'RJ']. None = todos os estados
    """
    data = df.copy()

    if estados is not None:
        data = data[data["Estado"].isin(estados)]

    mask = (
        (data["Gasto"] > 0) & (data["Votos"] > 0) & data["Situacao"].isin(SITUACAO_MAP)
    )
    data = data[mask].copy()
    data["Resultado"] = data["Situacao"].map(SITUACAO_MAP)

    cores_resultado = {"Eleito": CORES[2], "Suplente": CORES[0], "Não Eleito": CORES[3]}

    eleitos_gasto = data[data["Resultado"] == "Eleito"]["Gasto"]

    refs_verticais = {
        "Mín eleitos": (eleitos_gasto.min(), CORES[5], (4, 2)),
        "Mediana eleitos": (eleitos_gasto.median(), CORES[6], (8, 4)),
        "Média eleitos": (eleitos_gasto.mean(), CORES[7], (6, 3)),
        "Máx eleitos": (eleitos_gasto.max(), CORES[8], (2, 2)),
    }

    min_votos_media = data[data["Situacao"] == "ELEITO POR MÉDIA"]["Votos"].min()
    min_votos_qp = data[data["Situacao"] == "ELEITO POR QP"]["Votos"].min()

    titulo_extra = f" — {', '.join(estados)}" if estados else ""
    nome_arquivo = "gasto_vs_votos" + ("_" + "_".join(estados) if estados else "")

    fig, ax = plt.subplots(figsize=FIGSIZE)

    for resultado, grupo in data.groupby("Resultado"):
        ax.scatter(
            grupo["Gasto"],
            grupo["Votos"],
            c=cores_resultado[resultado],
            label=f"{resultado} (n={len(grupo)})",
            alpha=0.6,
            s=36,
            edgecolors="none",
        )

    for label, (valor, cor, dash) in refs_verticais.items():
        ax.axvline(
            valor,
            color=cor,
            linestyle="--",
            dashes=dash,
            label=f"{label}: R$ {valor:,.0f}",
        )

    if not pd.isna(min_votos_media):
        ax.axhline(
            min_votos_media,
            color=CORES[9],
            linestyle="-.",
            label=f"Mín votos eleito por média: {min_votos_media:,}",
        )
    if not pd.isna(min_votos_qp):
        ax.axhline(
            min_votos_qp,
            color=CORES[10],
            linestyle="-.",
            label=f"Mín votos eleito por QP: {min_votos_qp:,}",
        )

    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_xlabel("Gasto (R$) — escala log")
    ax.set_ylabel("Votos — escala log")
    ax.set_title(
        f"Gasto vs Votos por Situação Eleitoral — Deputado Federal 2014{titulo_extra}"
    )
    ax.legend(markerscale=2)
    plt.tight_layout()
    plt.savefig(f"assets/elections/plots/{nome_arquivo}.png")
    plt.show()
```

#### <font color=slategray> 1.3.2 - tabela_gasto_vs_votos()


**O que faz:** complementa o scatter com uma tabela de estatísticas (mediana, média, mínimo e máximo de gasto e votos) agregada por resultado eleitoral.

**Por que existe:**
- sempre usada junto com `plot_gasto_vs_votos`, mesma lógica de filtragem e agrupamento
- chamada 4 vezes (Brasil, RJ, GO, AM)
- o `titulo` é passado como argumento pois a numeração da tabela varia por seção

```python
def tabela_gasto_vs_votos(df, titulo, estados=None):
    """
    Args:
        df      : DataFrame — eleicoes_enriquecido ou qualquer subconjunto
        titulo  : str — ex: 'Tabela 3.6 - Gasto e Votos por Resultado Eleitoral'
        estados : list[str] ou None — ex: ['SP'], ['SP', 'RJ']. None = todos os estados
    """
    data = df.copy()

    if estados is not None:
        data = data[data["Estado"].isin(estados)]

    mask = (
        (data["Gasto"] > 0) & (data["Votos"] > 0) & data["Situacao"].isin(SITUACAO_MAP)
    )
    data = data[mask].copy()
    data["Resultado"] = data["Situacao"].map(SITUACAO_MAP)

    ordem = ["Eleito", "Suplente", "Não Eleito"]

    tabela = (
        data.groupby("Resultado")
        .agg(
            N=("Gasto", "count"),
            Gasto_Media=("Gasto", "mean"),
            Gasto_Mediana=("Gasto", "median"),
            Gasto_Min=("Gasto", "min"),
            Gasto_Max=("Gasto", "max"),
            Votos_Media=("Votos", "mean"),
            Votos_Mediana=("Votos", "median"),
        )
        .reindex(ordem)
    )

    total = pd.Series(
        {
            "N": data["Gasto"].count(),
            "Gasto_Media": data["Gasto"].mean(),
            "Gasto_Mediana": data["Gasto"].median(),
            "Gasto_Min": data["Gasto"].min(),
            "Gasto_Max": data["Gasto"].max(),
            "Votos_Media": data["Votos"].mean(),
            "Votos_Mediana": data["Votos"].median(),
        },
        name="Total",
    )
    tabela = pd.concat([tabela, total.to_frame().T])

    tabela["N"] = tabela["N"].map("{:,.0f}".format)
    for col in ["Gasto_Media", "Gasto_Mediana", "Gasto_Min", "Gasto_Max"]:
        tabela[col] = tabela[col].map("R$ {:,.0f}".format)
    for col in ["Votos_Media", "Votos_Mediana"]:
        tabela[col] = tabela[col].map("{:,.0f}".format)

    tabela.columns = [
        "N",
        "Gasto Médio",
        "Gasto Mediana",
        "Gasto Mín",
        "Gasto Máx",
        "Votos Médios",
        "Votos Mediana",
    ]

    print(titulo)
    display(tabela)
```

#### <font color=slategray> 1.3.3 - plot_barh()


**O que faz:** gráfico de barras horizontal com rótulos de valor, `textwrap` nos labels do eixo Y, `xlim` para evitar corte e salvamento das imgs padronizado.

**Por que existe:**
- padrão repetido 18 vezes no notebook
- o `textwrap` e o `xlim` corrigem problemas de overflow que aparecem quando os labels são longos

```python
def plot_barh(labels, values, titulo, xlabel, fmt, filename, cor=CORES[0]):
    """
    Args:
        labels   : Series ou list — labels do eixo Y
        values   : array — valores já convertidos (ex: / 1e6)
        titulo   : str — título do gráfico
        xlabel   : str — rótulo do eixo X
        fmt      : str — formato do bar_label (ex: "R$ %.1fM")
        filename : str — nome do arquivo sem extensão
        cor      : str — cor das barras (default CORES[0])
    """
    labels_w = [textwrap.fill(str(s), width=38) for s in labels]
    fig, ax = plt.subplots(figsize=FIGSIZE)
    bars = ax.barh(labels_w, values, color=cor)
    ax.set_xlabel(xlabel)
    ax.set_title(titulo)
    ax.bar_label(bars, fmt=fmt, padding=4)
    ax.set_xlim(0, max(values) * 1.4)
    plt.tight_layout()
    plt.savefig(f"assets/elections/plots/{filename}.png")
    plt.show()
```

#### <font color=slategray> 1.3.4 - plot_histograma()

**O que faz:** histograma com escala logarítmica no eixo X, linhas tracejadas de mediana e média com legenda e salvamento padronizado.

**Por que existe:**
- padrão repetido 7 vezes: histograma de gasto (Brasil, RJ, GO, AM) e de votos (RJ, GO, AM)
- a escala log é obrigatória por conta da assimetria extrema das distribuições

```python
def plot_histograma(
    data, n_bins, titulo, xlabel, filename, cor=CORES[0], prefixo_label=""
):
    """
    Args:
        data          : Series — valores filtrados (gasto > 0 ou votos > 0)
        n_bins        : int    — número de bins (raiz quadrada do n)
        titulo        : str    — título do gráfico
        xlabel        : str    — rótulo do eixo X
        filename      : str    — nome do arquivo sem extensão
        cor           : str    — cor das barras (default CORES[0])
        prefixo_label : str    — prefixo nas linhas de média/mediana (ex: "R$ ")
    """
    fig, ax = plt.subplots(figsize=FIGSIZE)
    ax.hist(data, bins=n_bins, color=cor, edgecolor="white")
    ax.set_xscale("log")
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Número de candidatos")
    ax.set_title(titulo)
    ax.axvline(
        data.median(),
        color=CORES[1],
        linestyle="--",
        label=f"Mediana: {prefixo_label}{data.median():,.0f}",
    )
    ax.axvline(
        data.mean(),
        color=CORES[3],
        linestyle="--",
        label=f"Média:   {prefixo_label}{data.mean():,.0f}",
    )
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"assets/elections/plots/{filename}.png")
    plt.show()
```

#### <font color=slategray> 1.3.5 - calc_bins()

**O que faz:** calcula `n_bins` pela regra da raiz quadrada (`int(√n)`), exibe uma tabela com os parâmetros do histograma (n, bins, mínimo, máximo, largura de bin) e retorna o valor calculado.

**Por que existe:**
- lógica repetida 7 vezes, sempre antes de `plot_histograma`
- retornar `n_bins` mantém o valor em escopo para a chamada do histograma na célula seguinte

```python
def calc_bins(data, titulo_tabela, rotulo, prefixo=""):
    """
    Args:
        data          : Series — valores filtrados (> 0)
        titulo_tabela : str    — ex: "Tabela 3.2 - Parâmetros do histograma"
        rotulo        : str    — "gasto" ou "votos" — usado nos rótulos da tabela
        prefixo       : str    — "R$ " para gasto, "" para votos
    Returns:
        n_bins : int
    """
    n_bins = int(np.sqrt(len(data)))
    largura = (data.max() - data.min()) / n_bins
    fmt = (lambda v: f"{prefixo}{v:,.2f}") if prefixo else (lambda v: f"{v:,.0f}")

    resumo = pd.DataFrame(
        {
            "Parâmetro": [
                f"Candidatos com {rotulo} > 0",
                "Número de bins (√n)",
                f"{rotulo.capitalize()} mínimo",
                f"{rotulo.capitalize()} máximo",
                "Largura de cada bin (linear)",
            ],
            "Valor": [
                f"{len(data):,}",
                f"{n_bins}",
                fmt(data.min()),
                fmt(data.max()),
                fmt(largura),
            ],
        }
    )
    print(titulo_tabela)
    display(resumo.set_index("Parâmetro"))
    return n_bins
```

#### <font color=slategray> 1.3.6 - plot_pearson()

**O que faz:** filtra candidatos com gasto e votos positivos, calcula o coeficiente de Pearson, exibe a tabela com `r` e p-valor, e plota a matriz de correlação como heatmap.

**Por que existe:**
- par tabela + heatmap repetido 4 vezes (Brasil, RJ, GO, AM)


```python
def plot_pearson(df, rotulo, titulo_tabela, filename):
    """
    Args:
        df            : DataFrame — eleicoes filtrado por estado (ou nacional)
        rotulo        : str       — ex: "RJ", "Brasil"
        titulo_tabela : str       — ex: "Tabela 4.6.1 — Correlação de Pearson — RJ"
        filename      : str       — nome do arquivo sem extensão
    """
    dados = df[["Gasto", "Votos"]].dropna()
    dados = dados[(dados["Gasto"] > 0) & (dados["Votos"] > 0)]

    r, p = pearsonr(dados["Gasto"], dados["Votos"])
    resultado = pd.Series(
        {"Coeficiente de Pearson (r)": round(r, 3), "p-valor": round(p, 4)},
        name=f"Gasto × Votos — {rotulo}",
    )
    print(titulo_tabela)
    display(resultado.to_frame())

    corr = dados[["Gasto", "Votos"]].corr()
    fig, ax = plt.subplots(figsize=(4, 3))
    im = ax.imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)
    plt.colorbar(im, ax=ax)
    ax.set_xticks(range(len(corr.columns)), corr.columns)
    ax.set_yticks(range(len(corr.index)), corr.index)
    for i in range(len(corr)):
        for j in range(len(corr.columns)):
            ax.text(
                j, i, f"{corr.iloc[i, j]:.3f}", ha="center", va="center", color="black"
            )
    ax.set_title(f"Matriz de correlação — {rotulo}")
    plt.tight_layout()
    plt.savefig(f"assets/elections/plots/{filename}.png")
    plt.show()
```

### <font color=steelblue> 2.3 - Validação dos dados mesclados

**Observações:**

apos o enriquecimento, o dataset possui duas colunas de gasto: `Gasto` (fornecida pelo professor) e `Total_Despesas_TSE` (calculada a partir da soma das despesas declaradas ao TSE). a validacao e feita em duas frentes:
- completude: todos os candidatos do dataset original estao presentes no TSE?
- consistencia: os valores de `Gasto` e `Total_Despesas_TSE` sao compativeis?

#### <font color=slategray> 2.3.1 - Completude

cruzei todos os candidatos do dataset original com o arquivo de prestacao de contas do TSE. candidatos sem match na coluna `Nome` sao aqueles sem nenhuma nota fiscal registrada. candidatos com `Gasto = 0` sem match sao esperados. candidatos com `Gasto > 0` sem match sao inconsistencias.

```python
sem_match = eleicoes_enriquecido["Nome"].isna()
sem_match_zero = sem_match & (eleicoes_enriquecido["Gasto"] == 0)
sem_match_valor = sem_match & (eleicoes_enriquecido["Gasto"] > 0)
encontrados = ~sem_match

resumo_completude = pd.DataFrame(
    {
        "Situacao": [
            "Total de candidatos no dataset original",
            "Encontrados nos dados TSE 2014",
            "Nao encontrados com Gasto = R$ 0 (sem despesas declaradas)",
            "Nao encontrados com Gasto > 0 (inconsistencias)",
        ],
        "Quantidade": [
            len(eleicoes_enriquecido),
            encontrados.sum(),
            sem_match_zero.sum(),
            sem_match_valor.sum(),
        ],
    }
)
resumo_completude["% do Total"] = (
    resumo_completude["Quantidade"] / len(eleicoes_enriquecido) * 100
).map("{:.2f}%".format)
resumo_completude["Quantidade"] = resumo_completude["Quantidade"].map("{:,}".format)

print("Tabela 2.3.1 - Completude do cruzamento entre eleicoes.csv e TSE 2014")
display(resumo_completude.set_index("Situacao"))
```

**Observações:**
- candidatos com `Gasto = 0` sem match no TSE sao esperados: o TSE so gera registro quando ha ao menos uma nota fiscal
- os poucos casos com `Gasto > 0` sem match representam inconsistencias pontuais e serao detalhados na secao 2.3.2
- apenas 4 candidatos não foram encontrados no dataset do TSE

#### <font color=slategray> 2.3.2 - Consistência dos valores

para os candidatos encontrados em ambos os datasets com `Gasto > 0`, comparei `Gasto` (eleicoes.csv) com `Total_Despesas_TSE` (soma das notas do TSE). classifico as diferencas em tres categorias: 
1. match exato  `< R$ 0.01`
2. diferenca residual de arredondamento `>= R$ 0.01` & `< R$ 1.00`
3. divergencia real. `>= R$ 1.00`

```python
# candidatos com Gasto > 0 sem match no TSE
nao_encontrados = eleicoes_enriquecido[sem_match_valor][
    ["Estado", "Numero Candidato", "Gasto"]
].copy()
nao_encontrados["Total_Despesas_TSE"] = 0.0
nao_encontrados["Diferenca"] = nao_encontrados["Gasto"]

print("Tabela 2.3.2a - Candidatos com Gasto > 0 nao encontrados no TSE")
display(nao_encontrados.reset_index(drop=True))

# consistencia para os candidatos cruzados com Gasto > 0
mask_cruzados = eleicoes_enriquecido["Total_Despesas_TSE"].notna() & (
    eleicoes_enriquecido["Gasto"] > 0
)
comparacao = eleicoes_enriquecido[mask_cruzados].copy()
comparacao["Diferenca"] = (comparacao["Gasto"] - comparacao["Total_Despesas_TSE"]).abs()

mask_match = comparacao["Diferenca"] < 0.01  # match exato
mask_residual = (comparacao["Diferenca"] >= 0.01) & (
    comparacao["Diferenca"] < 1
)  # match quase exato
mask_diverge = comparacao["Diferenca"] >= 1

resumo_consistencia = pd.DataFrame(
    {
        "Situacao": [
            "Match exato (diferenca < R$ 0,01)",
            "Diferenca residual (R$ 0,01 a R$ 1,00)",
            "Divergencia de valor (> R$ 1,00)",
            "Total candidatos cruzados",
        ],
        "Quantidade": [
            mask_match.sum(),
            mask_residual.sum(),
            mask_diverge.sum(),
            len(comparacao),
        ],
    }
)

resumo_consistencia["% do Cruzado"] = (
    resumo_consistencia["Quantidade"] / len(comparacao) * 100
).map("{:.2f}%".format)

resumo_consistencia["Quantidade"] = resumo_consistencia["Quantidade"].map("{:,}".format)

print("Tabela 2.3.2b - Consistencia entre Gasto (eleicoes.csv) e Total_Despesas_TSE")
display(resumo_consistencia.set_index("Situacao"))

print(f"\nDiferenca maxima: R$ {comparacao['Diferenca'].max():,.2f}")
print(f"Diferenca media:  R$ {comparacao['Diferenca'].mean():,.4f}")
```

**Observações:**
- a soma das despesas do TSE reproduz os valores do dataset original na grande maioria dos casos
- diferencas residuais (< R$ 1,00) sao atribuiveis a arredondamento de ponto flutuante
- `eleicoes.csv` e derivado do campo `Valor despesa` da prestacao de contas do TSE
- divergencias acima de R$ 1,00 sugerem que `eleicoes.csv` foi compilado a partir de uma versao anterior da prestacao de contas, anterior a revisao final pelo TSE
- a diferenca de versao nao compromete a analise: `Gasto` permanece como a medida oficial do gasto declarado

### <font color=steelblue> 3.8 - Correlação por Estado — Brasil inteiro

**Observações:**

```python
correlacoes_uf = []
for uf, grupo in eleicoes_enriquecido.groupby("Estado"):
    dados_uf = grupo[["Gasto", "Votos"]].dropna()
    dados_uf = dados_uf[(dados_uf["Gasto"] > 0) & (dados_uf["Votos"] > 0)]
    if len(dados_uf) >= 2:
        r_uf, p_uf = pearsonr(dados_uf["Gasto"], dados_uf["Votos"])
        correlacoes_uf.append(
            {
                "UF": uf,
                "r (Pearson)": round(r_uf, 3),
                "p-valor": round(p_uf, 4),
                "n (candidatos)": len(dados_uf),
            }
        )

tabela_corr_uf = (
    pd.DataFrame(correlacoes_uf)
    .sort_values("r (Pearson)", ascending=False)
    .reset_index(drop=True)
)
print("Tabela 3.8 — Coeficiente de Pearson entre Gasto e Votos por Estado")
display(tabela_corr_uf)
```

- estados grandes como SP e RJ apresentam correlações moderadas, refletindo maior heterogeneidade interna entre candidatos

### <font color=steelblue> 3.9 - Medidas de tendência central e dispersão por Estado

**Observações:**

tabela-resumo com média, mediana, desvio padrão, mínimo, máximo, Q1 e Q3 de votos e gasto para cada UF. estados nas linhas, medidas nas colunas. candidatos com gasto ou votos = 0 foram excluídos para não distorcer as estatísticas.

```python
def resumo_por_estado(coluna, mask):
    return (
        eleicoes_enriquecido[mask]
        .groupby("Estado")[coluna]
        .agg(
            Media="mean",
            Mediana="median",
            Desvio_Padrao="std",
            Minimo="min",
            Maximo="max",
            Q1=lambda x: x.quantile(0.25),
            Q3=lambda x: x.quantile(0.75),
        )
        .round(0)
        .astype(int)
    )


mask_votos = eleicoes_enriquecido["Votos"] > 0
mask_gasto = eleicoes_enriquecido["Gasto"] > 0

stats_votos = resumo_por_estado("Votos", mask_votos)
stats_gasto = resumo_por_estado("Gasto", mask_gasto)

tabela_resumo = pd.concat(
    {"Votos": stats_votos, "Gasto (R$)": stats_gasto},
    axis=1,
)

print("Tabela 3.9 — Medidas de tendência central e dispersão por Estado")
display(tabela_resumo)
```

**Observações:**
